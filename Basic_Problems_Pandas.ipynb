```python
# create and display a one-dimensional array-like object containing an array of data using Pandas module. 

import pandas as pd
data = [1, 2, 3, 4, 5]
s = pd.Series(data)
print(s)
```

```python
# Convert a Panda module Series to Python list and it's type.

a = [1, 2, 3, 4, 5]
s = pd.Series(a)
print("Pandas series \n", s)
print("Type", type(s))
print(s.tolist())
print(type(s.tolist()))
```

```python
# Add, subtract, multiply and divide two Pandas Series

a = pd.Series([2, 4, 6, 8, 10])
b = pd.Series([1, 3, 5, 7, 9])
print("Sum is \n", a + b)
print("Subtraction is\n", a - b)
print("Multiplication is\n", a * b)
print("Division is\n", a / b)
```

```python
# get the powers of an array values element-wise

a = pd.Series([ 0, 1, 2, 3, 4, 5, 6])
print("Original array\n", a)
power = a.pow(a)
print("Element wise power array\n", power)
```

```python
# Append a new row 'k' to data frame with given values for each column. Now delete the new row and return the original DataFrame

exam_data = {
    'Name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'], 
'Score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19], 
'Attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], 
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']} 
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'] 
df = pd.DataFrame(exam_data,index=labels)
print("Appending new row:")
df.loc['k'] = ['Suresh', 15.5, 1, 'yes']
print(df)
print('\nDelete the row and display the orginal dataframe')
df = df.drop('k')
print(df)
```

```python
#sort the DataFrame first by 'name' in descending order, then by 'score' in ascending order. 
df1 = df.sort_values(by = 'Name', ascending = False)
print("Sorting by names in descending order\n ", df1)
df1 = df.sort_values(by = 'Score')
print("\nSorting by scores in ascending order \n", df1)
```
