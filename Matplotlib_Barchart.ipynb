```python
# create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.

import numpy as np 
x = ['Group A','Group B','Group C','Group D', 'Group E']
M = [22, 30, 35, 35, 26]
W= [25, 32, 30, 35, 29]
X_axis = np.arange(len(x))
plt.bar(X_axis - 0.2, M, 0.4, label = 'Men')
plt.bar(X_axis + 0.2, W, 0.4, label = 'Women')
plt.xlabel("Groups")
plt.ylabel("Number of Employees")
plt.title("Number of Employees in each group")
plt.legend()
plt.show()
```

```python
# create bar plots with error bars on the same figure.
m = [1,2,3,4]
v = [ 0.2474, 0.1235, 0.1737, 0.1824 ]
sd = [0.3314, 0.2278, 0.2836, 0.2645 ]
plt.bar(m,v)
plt.errorbar(m,v,yerr=sd,fmt="o",color='red')
plt.show()
```

```python
# create a stacked bar plot with error bars.
x = ['Group A','Group B','Group C','Group D','Group E']
men = (22, 30, 35, 35, 26)
women = (25, 32, 30, 35, 29)
menSd = (4, 3, 4, 1, 5)
womenSd = (3, 5, 2, 3, 3)
x_axis = np.arange(len(x))    
plot1 = plt.bar(x_axis, men, width=0.35, yerr=menSd)
plot2 = plt.bar(x_axis, women, width=0.35, bottom=men, yerr=womenSd)
plt.ylabel('Scores')
plt.xlabel('Groups')
plt.title('Scores by group and gender')
plt.xticks(x_axis, x)
plt.legend((plot1[0], plot2[0]), ('Men', 'Women'))
plt.show()
```

```python
#  create a horizontal bar chart with differently ordered colors. 
import numpy as np
import matplotlib.pyplot as plt
students = ('student1','student2','student3')
languages    = [['Language','Science','Math'], 
               ['Science','Math','Language'], 
               ['Math','Language','Science']]
numbers = [{'Language':75, 'Science':88, 'Math':96},
           {'Language':71, 'Science':95, 'Math':92},
           {'Language':75, 'Science':90, 'Math':89}]
order = np.arange(3)
language=np.array([75,71,75])
Math=np.array([96,92,89])
Science=np.array([71,95,90])
fig = plt.subplots(figsize =(7, 5))
width=0.30
p1 =plt.barh(order,language,width,color='r')
p2 =plt.barh(order,Math,width,left=language,color='g')
p3 =plt.barh(order,Science,width,left=language+Math,color='b')
plt.yticks(order, students)
plt.legend((p1[0],p2[0],p3[0]),('Language','Math','Science'),loc="best", bbox_to_anchor=(1.0, 1.00))
plt.show()
```

```python
#  create stack bar plot and add label to each section
import matplotlib.pyplot as plt
people = ('G1','G2','G3','G4','G5','G6','G7','G8') 
data = [[ 3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039, 7.1123587, 12.77792868, 3.44773477], 
[ 11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195, 6.79685302, 7.24578743, 3.69371847], 
[ 3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182, 4.63832778, 11.16849999, 8.56883433], 
[ 4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185, 10.97567589, 3.98287652, 8.80552122]] 
plt.figure(figsize=(8,6))
stack = np.zeros(len(people))
for row in range(len(data)):
    text = plt.bar(people,data[row], bottom = stack)
    stack = stack + data[row]
    plt.bar_label(text, label_type = 'center')
plt.show()
```
