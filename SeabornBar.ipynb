```python
# Draw bar plot of sex against survived for a dataset given in the url 
import seaborn as sns
import matplotlib.pyplot as plt
titanic_data = sns.load_dataset("titanic")
sns.barplot(x = "sex", y = "survived", data = titanic_data)
plt.show()
```

```python
titanic_data = sns.load_dataset("titanic")
sns.countplot(x='deck',data = titanic_data)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
sns.pointplot(x="sex",y="survived",data=titanic)
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="day",y="total_bill")
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
species = sns.load_dataset("iris")
sns.violinplot(x="species", y="sepal_length", data=species)
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="tip", data=tips)
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.swarmplot(x="size", y="total_bill", data=tips,size=3)
plt.show()
```
