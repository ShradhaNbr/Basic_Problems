```python
# create a pie chart of the popularity of programming Languages. 
import matplotlib.pyplot as plt
lang = ['Java', 'Python', 'PHP', 'JavaScript', 'C#',  'C++']
pop = [22.2, 17.6, 8.8, 8, 7.7, 6.7 ]
fig = plt.figure(figsize =(5, 5))
plt.pie(pop,  labels = lang)
plt.show()
```

```python
#create a pie chart with a title of the popularity of programming Languages. 
lang = ['Java', 'Python', 'PHP', 'JavaScript', 'C#',  'C++']
pop = [22.2, 17.6, 8.8, 8, 7.7, 6.7 ]
fig = plt.figure(figsize =(5, 5))
plt.pie(pop, labels = lang)
plt.title(label = 'Popularity of programming languages')
plt.show()
```

```python
# create a pie chart with a title of the popularity of programming Languages.Make multiple wedges of the pie. 
lang = ['Java', 'Python', 'PHP', 'JavaScript', 'C#',  'C++']
pop = [22.2, 17.6, 8.8, 8, 7.7, 6.7 ]
explode = (0.1, 0.1, 0, 0, 0,0.1)
plt.pie(pop, explode = explode, labels = lang, autopct = '%1.1f%%')
plt.title(label = 'Popularity of programming languages')
plt.show()
```
